```python
import pandas as pd

check = pd.read_csv('analysis_tablet10.csv')
```

```python
# 1. 'ìƒí˜¸ëª…' ì»¬ëŸ¼ ì‚­ì œ (ì´ì „ ìš”ì²­ ì‚¬í•­)
check.drop(columns=['ìƒí˜¸ëª…'], inplace=True)
```

```python
# 2. í–‰ì •ë™ ê¸°ì¤€ ì¹´í˜ ë°€ë„ ê³„ì‚°
# transform('count')ëŠ” ê° í–‰ì´ ì†í•œ ë™ì— ì´ ëª‡ ê°œì˜ ë°ì´í„°ê°€ ìˆëŠ”ì§€ ê³„ì‚°í•´ì„œ ì±„ì›Œì¤ë‹ˆë‹¤.
check['ì¹´í˜_ë°€ë„'] = check.groupby('í–‰ì •ë™ëª…')['brand'].transform('count')
```

```python
# 3. ì—­ ê±°ë¦¬ ì´ë¶„í™” (500m ê¸°ì¤€)
# lambda í•¨ìˆ˜ë¥¼ ì‚¬ìš©í•˜ì—¬ 500m ì´í•˜ì¸ ê²½ìš° 'ê·¼ì ‘', ê·¸ ì™¸ëŠ” 'ë¹„ê·¼ì ‘'ìœ¼ë¡œ ë¶„ë¥˜í•©ë‹ˆë‹¤.
check['ì—­_ì ‘ê·¼ì„±'] = check['ì—­_ê±°ë¦¬'].apply(lambda x: 'ê·¼ì ‘' if x <= 500 else 'ë¹„ê·¼ì ‘')

# ê²°ê³¼ í™•ì¸
print(check[['í–‰ì •ë™ëª…', 'ì¹´í˜_ë°€ë„', 'ì—­_ê±°ë¦¬', 'ì—­_ì ‘ê·¼ì„±']].head())
```

```python
# 5. í”„ëœì°¨ì´ì¦ˆ(ë¸Œëœë“œ) ë¹„ìœ¨ ê³„ì‚° ğŸ¢
# í–‰ì •ë™ë³„ë¡œ 'íˆ¬ë‹¤ë¦¬'ì™€ 'ì—­ì „í• ë¨¸ë‹ˆë§¥ì£¼'ì˜ ë¹„ì¤‘ì„ ê°ê° êµ¬í•©ë‹ˆë‹¤.
brand_counts = check.groupby(['í–‰ì •ë™ëª…', 'brand']).size().unstack(fill_value=0)
brand_total = brand_counts.sum(axis=1)
```

```python
# 'columns' ì˜µì…˜ì— {ê¸°ì¡´ì´ë¦„ : ìƒˆì´ë¦„} í˜•íƒœë¡œ ì ì–´ì¤ë‹ˆë‹¤.
# inplace=TrueëŠ” ë³€ê²½ ì‚¬í•­ì„ ë°ì´í„°í”„ë ˆì„ì— ì¦‰ì‹œ ë°˜ì˜í•˜ë¼ëŠ” ëœ»ì…ë‹ˆë‹¤.
check.rename(columns={'ì¹´í˜_ë°€ë„': 'ì£¼ì _ë°€ë„'}, inplace=True)

# ì˜ ë°”ë€Œì—ˆëŠ”ì§€ ìƒë‹¨ 5ê°œ í–‰ì„ í™•ì¸í•©ë‹ˆë‹¤.
print(check[['í–‰ì •ë™ëª…', 'ì£¼ì _ë°€ë„']].head())
```

```python
# 1. í–‰ì •ë™ë³„ë¡œ ê° ë¸Œëœë“œê°€ ëª‡ ê°œì”© ìˆëŠ”ì§€ í‘œë¥¼ ë§Œë“­ë‹ˆë‹¤.
brand_counts = check.groupby(['í–‰ì •ë™ëª…', 'brand']).size().unstack(fill_value=0)

# 2. í–‰ì •ë™ë³„ ì „ì²´ ë§¤ì¥ ìˆ˜ë¥¼ êµ¬í•©ë‹ˆë‹¤.
total_counts = brand_counts.sum(axis=1)

# 3. 'íˆ¬ë‹¤ë¦¬'ì™€ 'ì—­ì „í• ë¨¸ë‹ˆ'ì˜ ë¹„ìœ¨ì„ ê³„ì‚°í•˜ì—¬ ìƒˆë¡œìš´ ì»¬ëŸ¼ìœ¼ë¡œ ì¶”ê°€í•©ë‹ˆë‹¤.
check['íˆ¬ë‹¤ë¦¬_ë¹„ìœ¨'] = check['í–‰ì •ë™ëª…'].map(brand_counts['íˆ¬ë‹¤ë¦¬'] / total_counts)
check['ì—­ì „í• ë¨¸ë‹ˆ_ë¹„ìœ¨'] = check['í–‰ì •ë™ëª…'].map(brand_counts['ì—­ì „í• ë¨¸ë‹ˆ'] / total_counts)

# ê²°ê³¼ í™•ì¸
print(check[['í–‰ì •ë™ëª…', 'íˆ¬ë‹¤ë¦¬_ë¹„ìœ¨', 'ì—­ì „í• ë¨¸ë‹ˆ_ë¹„ìœ¨']].head())
```
