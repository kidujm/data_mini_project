```python
import pandas as pd

place_data =  pd.read_csv('place_data_tudari.csv') 
place_data
```

```python
# brand 컬럼에서 '투다리'만 필터링
place_data = place_data[place_data["brand"] == "투다리"]

# 결과 확인
place_data.head()

# CSV로 저장
place_data.to_csv(
    "place_data_tudari.csv",
    index=False,
    encoding="utf-8-sig"
)
```

```python
place_data.head()
place_data.columns
place_data = place_data.dropna(subset=['place_id']) #결측값 제거

```

```python
# 네이버 플레이스 URL 생성
place_data['naver_place_url'] = (
    'https://pcmap.place.naver.com/place/'
    + place_data['place_id'].astype(str)
    + '/home'
)
# 결과 확인
place_data[['place_id', 'naver_place_url']].head()
```

```python
# =====================================
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
import time
import re
import pandas as pd

options = Options()
options.add_experimental_option("detach", True)

service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=options)

wait = WebDriverWait(driver, 15)
```

```python
# 한페이지만 출력
driver.get(place_data['naver_place_url'].iloc[0])
```

```python

# =====================================
# naver_place_url 없으면 생성 (안전)
# =====================================
if "naver_place_url" not in place_data.columns:
    place_data["naver_place_url"] = (
        "https://pcmap.place.naver.com/place/"
        + place_data["place_id"].astype(str)
        + "/home"
    )

# =====================================
# 숫자만 추출 함수
# =====================================
def extract_number(text):
    return int(re.sub(r"[^0-9]", "", text))


# =====================================
# 결과 저장용 리스트
# =====================================
results = []
fail_place_ids = []


# =====================================
# 전체 place_data 반복 크롤링
# =====================================
for _, row in place_data[["place_id", "naver_place_url"]].dropna().head(10).iterrows():
    place_id = row["place_id"]
    url = row["naver_place_url"]

    try:
        # -------------------------------
        # 페이지 접속
        # -------------------------------
        driver.get(url)
        time.sleep(3)

        # (필요 시 iframe)
        # driver.switch_to.frame("entryIframe")

        # -------------------------------
        # 1️⃣ 근접 지하철역 (가장 먼저)
        # -------------------------------
        near_subway = wait.until(
            EC.presence_of_element_located(
                (By.XPATH,
                 '//*[@id="app-root"]/div/div/div[6]/div/div[2]/div[1]/div/div[1]/div/div')
            )
        ).text

        # -------------------------------
        # 2️⃣ 방문자 리뷰 수
        # -------------------------------
        visit_review_count = extract_number(
            wait.until(
                EC.presence_of_element_located(
                    (By.XPATH,
                     '//*[@id="app-root"]/div/div/div[2]/div[1]/div[2]/span[2]/a')
                )
            ).text
        )

        # -------------------------------
        # 3️⃣ 블로그 리뷰 수
        # -------------------------------
        blog_review_count = extract_number(
            wait.until(
                EC.presence_of_element_located(
                    (By.XPATH,
                     '//*[@id="app-root"]/div/div/div[2]/div[1]/div[2]/span[3]/a')
                )
            ).text
        )

        # -------------------------------
        # 4️⃣ 리뷰 탭 클릭
        # -------------------------------
        review_button = wait.until(
            EC.element_to_be_clickable(
                (By.XPATH,
                 '//*[@id="app-root"]/div/div/div[5]/div/div/div/div/a[3]/span')
            )
        )
        driver.execute_script("arguments[0].click();", review_button)
        time.sleep(2)

        # -------------------------------
        # 스크롤 1회
        # -------------------------------
        driver.execute_script("window.scrollBy(0, window.innerHeight);")
        time.sleep(2)

        # -------------------------------
        # 5️⃣ 사진 리뷰 버튼 클릭
        # -------------------------------
        photo_button = wait.until(
            EC.element_to_be_clickable(
                (By.XPATH,
                 '//*[@id="app-root"]/div/div/div[7]/div[3]/div[3]/h2/div[2]/a')
            )
        )
        driver.execute_script("arguments[0].click();", photo_button)
        time.sleep(2)

        # -------------------------------
        # 6️⃣ 총 리뷰 수
        # -------------------------------
        total_review_count = extract_number(
            wait.until(
                EC.presence_of_element_located(
                    (By.XPATH,
                     '//*[@id="app-root"]/div/div/div[7]/div[3]/div[3]/h2/div[1]/em')
                )
            ).text
        )

        # -------------------------------
        # 7️⃣ 사진 포함 리뷰 수
        # -------------------------------
        photo_review_count = extract_number(
            wait.until(
                EC.presence_of_element_located(
                    (By.XPATH,
                     '//*[@id="app-root"]/div/div/div[7]/div[3]/div[3]/h2/div[1]/em')
                )
            ).text
        )

        # -------------------------------
        # 결과 누적
        # -------------------------------
        results.append({
            "place_id": place_id,
            "naver_place_url": url,
            "near_subway": near_subway,
            "visit_review_count": visit_review_count,
            "blog_review_count": blog_review_count,
            "total_review_count": total_review_count,
            "photo_review_count": photo_review_count
        })

        print(f"[OK] 수집 완료 | place_id={place_id}")

    except Exception as e:
        print(f"[ERROR] 수집 실패 | place_id={place_id}")
        print(e)
        fail_place_ids.append(place_id)
        continue


# =====================================
# DataFrame 변환
# =====================================
df = pd.DataFrame(results)
df
```
